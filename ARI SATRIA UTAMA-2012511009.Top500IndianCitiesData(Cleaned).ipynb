data = pd.read_csv('/content/drive/MyDrive/indian_cities.csv')

```python
top_5_states_population = data.groupby('State')['Population'].sum().nlargest(5).reset_index()
top_5_states_population.columns = ['State', 'Total Population']
display(top_5_states_population)
```

tabel diatas menjelaskan 5 state dengan populasi tertinggi

```python
# Create a bar plot for top 5 states by population
plt.figure(figsize=(12, 7))
sns.barplot(x='State', y='Total Population', data=top_5_states_population, palette='magma')
plt.title('Top 5 Negara Bagian dengan Populasi Terpadat')
plt.xlabel('Negara Bagian')
plt.ylabel('Total Populasi')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

diagram diatas menjelaskan 5 state dengan populasi terpadat yakni
1. maharashtra
2. Uttar Pradesh
3. Andhra Pradesh
4. West Bengal
5. Gujarat

```python
# Sort the DataFrame by 'Population' in descending order and select the top 5 cities
top_5_cities_population = data.sort_values(by='Population', ascending=False).head(5)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Population', data=top_5_cities_population, palette='viridis')
plt.title('Top 5 Kota Teratas Berdasarkan Populasi')
plt.xlabel('Kota')
plt.ylabel('Populasi')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

diagram diatas menjelaskan terkait 5 kota dengan populasi terpadat di india

```python
top_5_states_literacy = data.groupby('State')['Literacy'].mean().nlargest(5).reset_index()
top_5_states_literacy.columns = ['State', 'Average Literacy']
display(top_5_states_literacy)
```

tabel di atas menjelaskan 5 state (negara bagian) dengan tingkat literacy tertinggi

```python

# Create a bar plot for top 5 states by literacy
plt.figure(figsize=(12, 7))
sns.barplot(x='State', y='Average Literacy', data=top_5_states_literacy, palette='viridis')
plt.title('Top 5 Negara Bagian dengan Tingkat Literasi Tertinggi')
plt.xlabel('Negara Bagian')
plt.ylabel('Rata-rata Literasi')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(data, x='Population', y='Literacy', color='City', hover_name='City', title='Hubungan antara Populasi, Literasi, dan Kota')
fig.show()
```

diagram scater tersebut menjelaskan hubungan antara populasi literasi dengan kota yang ada di india

Adapun maksud dari kode diagram pencar Plotly Express yang diatas ialah:

*   **`fig = px.scatter(data, x='Population', y='Literacy', color='City', hover_name='City', title='Hubungan antara Populasi, Literasi, dan Kota')`**:
    *   `px.scatter()`: Ini adalah fungsi dari library `plotly.express` yang digunakan untuk membuat diagram pencar (scatter plot).
    *   `data`: Argumen ini menentukan DataFrame yang akan digunakan untuk plotting, yaitu `data` DataFrame Anda.
    *   `x='Population'`: Mengatur kolom 'Population' dari DataFrame sebagai sumbu horizontal (x-axis) diagram.
    *   `y='Literacy'`: Mengatur kolom 'Literacy' dari DataFrame sebagai sumbu vertikal (y-axis) diagram.
    *   `color='City'`: Menginstruksikan Plotly untuk mewarnai setiap titik data berdasarkan nilai unik di kolom 'City'. Ini membantu membedakan kota-kota yang berbeda dalam plot.
    *   `hover_name='City'`: Ketika Anda mengarahkan kursor ke titik mana pun di plot, nama kota (dari kolom 'City') akan ditampilkan sebagai informasi hover.
    *   `title='Hubungan antara Populasi, Literasi, dan Kota'`: Menetapkan judul untuk diagram pencar, yang menjelaskan tujuan plot ini.

*   **`fig.show()`**:
    *   Metode ini digunakan untuk menampilkan objek figur (`fig`) yang telah dibuat. Ini akan merender diagram pencar interaktif di output sel.
